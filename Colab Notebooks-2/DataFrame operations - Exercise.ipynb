```
import pandas as pd
df = pd.DataFrame([[18,10,5,11,-2],
                    [2,-2,9,-11,3],
                    [-4,6,-19,2,1],
                    [3,-14,1,-2,8],
                    [-2,2,4,6,13]],
                  index = list('pqrst'),
                    columns = list('abcde'))
          
```

```
import numpy as np
```

```
new_df = pd.DataFrame(df[df.apply(np.sum, axis = 1) % 2 == 0])
```

```
new_df  
```

```
new_df.to_excel('file_df.xlsx', sheet_name = 'Sheet1')
```

```
df_temp = new_df
```

```
df_temp
```

```
df_temp.eval('m = (a*b*c*d*e)', inplace=True)
```

```
df_temp
```

```
writer = pd.ExcelWriter('file_df.xlsx')
df_temp.to_excel(writer, sheet_name='Sheet2')
writer.save()
writer.close()
```
