```
 
import numpy as np
mat = np.array([[10,5,9],
                [2,20,6],
                [8,3,30]]).reshape(3,3)
```

```
N1 = mat[0].max()
N2 = mat[1].max()
N3 = mat[2].max()
```

```
mat[np.tril_indices(3, -1)] = mat[np.tril_indices(3, -1)] + N3
```

```
mat
```

```
mat[np.mask_indices(3, np.triu, 1)] = mat[np.mask_indices(3, np.triu, 1)] + N1
```

```
mat
```

```
mat[np.diag_indices(3)] = mat[np.diag_indices(3)] + N2
```

```
mat
```
