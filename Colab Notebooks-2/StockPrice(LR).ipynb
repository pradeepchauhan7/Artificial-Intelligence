```
import quandl
```

```
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
df = quandl.get("WIKI/AMZN")
```

```
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```
df = df.reset_index()
```

```
prices = df['Close'].tolist()
```

```
dates = df.index.tolist()
```

```
dates = np.reshape(dates, (len(dates), 1))
```

```
prices = np.reshape(prices, (len(prices), 1))
```

```
regressor = LinearRegression()
```

```
xtrain, xtest, ytrain, ytest = train_test_split(dates, prices, test_size = 0.33, random_state = 42)
regressor.fit(xtrain, ytrain)
```

```
plt.scatter(xtrain, ytrain, color = 'green', label = 'Actual Price', s = 1)
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue', linewidth = 3, label = 'Predicted Price')
plt.title('Linear Regression(Train set) | Time vs Price')
plt.legend()
plt.xlabel('Date Integer')
plt.ylabel('Price')
plt.show()
```

```
plt.scatter(xtest, ytest, color = 'green', label = 'Actual Price', s = 0.5)
plt.plot(xtest, regressor.predict(xtest), color = 'blue', linewidth = 1, label = 'Predicted Price')
plt.title('Linear Regression(Test set) | Time vs Price')
plt.legend()
plt.xlabel('Date Integer')
plt.ylabel('Price')
plt.show()
```
